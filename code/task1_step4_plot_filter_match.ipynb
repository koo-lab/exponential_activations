```python
import os
import numpy as np
from six.moves import cPickle
import matplotlib.pyplot as plt
%matplotlib inline
```

# Comparison plot of classification performance

```python
results_path = os.path.join('../results/', 'task1')
file_path = os.path.join(results_path, 'task1_performance_results.pickle')
with open(file_path, 'rb') as f:
    results = cPickle.load(f)
```

# Filter match comparison to ground truth motifs (Fig. 1b)

```python
results_path = os.path.join('../results/', 'task1')
file_path = os.path.join(results_path, 'task1_filter_results.pickle')
with open(file_path, 'rb') as f:
    results = cPickle.load(f)
```

```python

activations = ['relu', 'exponential', 'sigmoid', 'tanh', 'softplus', 'linear',  'elu']
activation_names = ['Relu', 'Exponential', 'Sigmoid', 'Tanh', 'Softplus', 'Linear', 'Elu']

fig = plt.figure(figsize=(7,8))
ax = plt.subplot(3,1,1)
vals = []
for activation in activations:
    vals.append(results['cnn-2'][activation]['match_fraction'])
ax.boxplot(vals);
ax.set_xticklabels([])
plt.ylabel('CNN-2', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);

ax = plt.subplot(3,1,2)
vals = []
for activation in activations:
    vals.append(results['cnn-50'][activation]['match_fraction'])
ax.boxplot(vals);
ax.set_xticklabels([])
plt.ylabel('CNN-50', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);

ax = plt.subplot(3,1,3)
vals = []
for activation in activations:
    vals.append(results['cnn-deep'][activation]['match_fraction'])
ax.boxplot(vals);
plt.ylabel('CNN-deep', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
plt.xticks(range(1,len(activations)+1), activation_names, fontsize=12, rotation=45);

outfile = os.path.join(results_path, 'task1_filter_match.pdf')
fig.savefig(outfile, format='pdf', dpi=200, bbox_inches='tight')

```

# Filter match comparison to ground truth motifs for modified activations (Extended data Fig. 1a)

```python
from scipy import stats

fig = plt.figure(figsize=(9,8))
ax = plt.subplot(3,4,1)
vals = []
vals.append(results['cnn-2']['sigmoid']['match_fraction'])
vals.append(results['cnn-2']['shift_scale_sigmoid']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
plt.ylabel('CNN-2', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-2']['sigmoid']['match_fraction'], 
                       results['cnn-2']['shift_scale_sigmoid']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,5)
vals = []
vals.append(results['cnn-50']['sigmoid']['match_fraction'])
vals.append(results['cnn-50']['shift_scale_sigmoid']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
plt.ylabel('CNN-50', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-50']['sigmoid']['match_fraction'], 
                       results['cnn-50']['shift_scale_sigmoid']['match_fraction']).pvalue
print(pval)


ax = plt.subplot(3,4,9)
vals = []
vals.append(results['cnn-deep']['sigmoid']['match_fraction'])
vals.append(results['cnn-deep']['shift_scale_sigmoid']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
plt.ylabel('CNN-deep', fontsize=12)
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
plt.xticks(range(1,3), ['Sigmoid', 'Modified-Sigmoid'], fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-deep']['sigmoid']['match_fraction'], 
                       results['cnn-deep']['shift_scale_sigmoid']['match_fraction']).pvalue
print(pval)


ax = plt.subplot(3,4,2)
vals = []
vals.append(results['cnn-2']['tanh']['match_fraction'])
vals.append(results['cnn-2']['shift_scale_tanh']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-2']['tanh']['match_fraction'], 
                       results['cnn-2']['shift_scale_tanh']['match_fraction']).pvalue
print(pval)


ax = plt.subplot(3,4,6)
vals = []
vals.append(results['cnn-50']['tanh']['match_fraction'])
vals.append(results['cnn-50']['shift_scale_tanh']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-50']['tanh']['match_fraction'], 
                       results['cnn-50']['shift_scale_tanh']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,10)
vals = []
vals.append(results['cnn-deep']['tanh']['match_fraction'])
vals.append(results['cnn-deep']['shift_scale_tanh']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
ax.tick_params(labelleft=False) 
plt.xticks(range(1,3), ['Tanh', 'Modified-Tanh'], fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-deep']['tanh']['match_fraction'], 
                       results['cnn-deep']['shift_scale_tanh']['match_fraction']).pvalue
print(pval)


ax = plt.subplot(3,4,3)
vals = []
vals.append(results['cnn-2']['relu']['match_fraction'])
vals.append(results['cnn-2']['shift_scale_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-2']['relu']['match_fraction'], 
                       results['cnn-2']['shift_scale_relu']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,7)
vals = []
vals.append(results['cnn-50']['relu']['match_fraction'])
vals.append(results['cnn-50']['shift_scale_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-50']['relu']['match_fraction'], 
                       results['cnn-50']['shift_scale_relu']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,11)
vals = []
vals.append(results['cnn-deep']['relu']['match_fraction'])
vals.append(results['cnn-deep']['shift_scale_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
ax.tick_params(labelleft=False) 
plt.xticks(range(1,3), ['Relu', 'Modified-Relu'], fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-deep']['relu']['match_fraction'], 
                       results['cnn-deep']['shift_scale_relu']['match_fraction']).pvalue
print(pval)


ax = plt.subplot(3,4,4)
vals = []
vals.append(results['cnn-2']['exponential']['match_fraction'])
vals.append(results['cnn-2']['exp_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-2']['exponential']['match_fraction'], 
                       results['cnn-2']['exp_relu']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,8)
vals = []
vals.append(results['cnn-50']['exponential']['match_fraction'])
vals.append(results['cnn-50']['exp_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
ax.set_xticklabels([])
ax.tick_params(labelleft=False) 
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
#plt.xticks(range(1,len(activations)+1), activations, fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-50']['exponential']['match_fraction'], 
                       results['cnn-50']['exp_relu']['match_fraction']).pvalue
print(pval)

ax = plt.subplot(3,4,12)
vals = []
vals.append(results['cnn-deep']['exponential']['match_fraction'])
vals.append(results['cnn-deep']['exp_relu']['match_fraction'])
ax.boxplot(vals,  widths = 0.6);
plt.yticks([0., .25, 0.5, 0.75, 1.0], fontsize=12)
ax.tick_params(labelleft=False) 
plt.xticks(range(1,3), ['Exp', 'Modified-Exp'], fontsize=12, rotation=45);
pval = stats.ttest_ind(results['cnn-deep']['exponential']['match_fraction'], 
                       results['cnn-deep']['exp_relu']['match_fraction']).pvalue
print(pval)

outfile = os.path.join(results_path, 'task1_filter_match_modified_activations.pdf')
fig.savefig(outfile, format='pdf', dpi=200, bbox_inches='tight')

```
